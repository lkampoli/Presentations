Monitor points are collected during runtime as files "point_i.hdf5" where i is the index of the monitor point. The structure of the file can be seen in figure below left. The group point info contains the information contained in the grey box which essentially consists of the flow information that is required to postprocess results and the position of point in space.

The group QPH_t_Series contains time series of various quantities collected, where nit and t are the iteration number and time at the collection points of qph. Using other options in hipstar the other time series like the veolicty gradient duidxi in this case can be added.

The hdf5 format allows to add other datasets and in this case span_specs are collected, separated in datasets for intervalls that can be uses for convergence studies. Using the "hdfview" command allows to open the hdf5 binary file, see the file structure and do some simple plotting.

Using the old input file in hipstar the monitor point specification is slightly changed. Previously it contained to starting integers 

a b 

where a is the number of points and b is the capturing intervall. This is now repaced with 5 integers

a b c d e 

where a and b are the same, c is the intervall at which the monitor points are dumped to file, d is the option and e is the index in the spanwise direction that is collected. The integers are effective for all points that are subsequently given in physical coordinates (-Dspan_spec) or grid coordinates if the -Dspan_spec flag is not used. Currently only options (given through integer d) 1 and 3 are tested where 1 dumps the primitive variables in physical space $ \rho,u,v,w,T$ and 3 dumps the same as one and in addition the velocity gradient tensor. With the new infile a new specification will enter.
