```python
mon_point_id=2
var_index=7
skip=10
save_fig=False
#======================================
#  end user input
#======================================

start_name='monitor_point_%i'%mon_point_id
files= os.listdir('../mon_point/')
red_files=[]
for f in files:
    if f.startswith(start_name):
        temp='../mon_point/'+f
        red_files.append(temp)

red_files.sort()
time=[]
step=[]
var=[]
for f in red_files:
    file1=open(f,'r')
    count=1
    for line in file1:
        if count==0:
            li=line.split()
            step.append(float(li[0]))
            time.append(float(li[1]))
            var.append(float(li[var_index]))
            count=10
        else:
            count-=count
plt.figure()
plt.plot(step,var)
plt.xlabel('timestep')
plt.ylabel('var')
if save_fig:
    plt.savefig('time_hist_var.pdf')
     
```
