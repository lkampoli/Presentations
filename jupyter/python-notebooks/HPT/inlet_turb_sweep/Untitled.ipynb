```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rotate(x,alpha):
    """
    rotate the vector x by the angle alpha (2d)
    """
    c_ang=np.cos(alpha*np.pi/180.)
    s_ang=np.sin(alpha*np.pi/180.)
    y=np.empty(2)
    y[0]=x[0]*c_ang-x[1]*s_ang
    y[1]=x[0]*s_ang+x[1]*c_ang
    return y
```

```python
x=[1,np.tan(30.*np.pi/180.)]
y=rotate(x,-30)
y1=rotate(x,90)
```

```python
plt.figure()
plt.axes(aspect=1.)
plt.plot([0,x[0]],[0,x[1]])
plt.plot([0,y[0]],[0,y[1]])
plt.plot([0,y1[0]],[0,y1[1]])
plt.grid()
```

```python
norm=np.array([0.5,1.])
l_n=np.sqrt(norm[0]**2+norm[1]**2)
norm/=l_n
tang=np.empty(2)
tang[0] = norm[1]
tang[1] = -norm[0]
ang=np.arctan(tang[1]/tang[0])
c_ang=np.cos(ang)
s_ang=np.sin(ang)
print 'ang=%7.3f'%(ang/np.pi*180.)
```

```python
u_v=np.array([2.,1.])
u_v_t=np.empty(2)
u_v_r=rotate(u_v,-ang/np.pi*180.)
u_v_t[0]=u_v[0]*tang[0]+u_v[1]*tang[1]
u_v_t[1]=u_v[0]*norm[0]+u_v[1]*norm[1]
```

```python
plt.figure()
plt.axes(aspect=1.)
plt.plot([0.,norm[0]],[0.,norm[1]])
plt.plot([0.,tang[0]],[0.,tang[1]])
```

```python
plt.figure()
plt.axes(aspect=1.)
plt.plot([0.,u_v_t[0]],[0.,u_v_t[1]],'--')
plt.plot([0.,u_v_r[0]],[0.,u_v_r[1]],'-.')
```
