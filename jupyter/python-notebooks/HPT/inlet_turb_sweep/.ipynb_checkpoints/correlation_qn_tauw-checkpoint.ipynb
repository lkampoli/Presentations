```python
%%time
blocks=[2,1,3,4]
n_bl=len(blocks)
start_TS=5400601
stop_TS= 5500101
inc_TS=500
steps=range(start_TS,stop_TS,inc_TS)
tau_w=lsub.subspace('WALL_SH_HEAT',20,start_TS,blocks,source_folder='./')
tau_w.load_grid()
tau_w.load_data(steps,read_all=True)
```

```python
plt.figure()
dist=[]
for ib in range(n_bl):
    nx,ny,nz=tau_w.header[0][ib]
    dist.append(np.empty(ny))
    if ib==0:
        dist[ib][0]=0.
    else:
        dist[ib][0]=dist[ib-1][-1]
    for j in range(1,ny):
        dist[ib][j]=dist[ib][j-1]+np.sqrt((tau_w.grid[ib][0,0,j,0]-tau_w.grid[ib][0,0,j-1,0])**2
                                         +(tau_w.grid[ib][1,0,j,0]-tau_w.grid[ib][1,0,j-1,0])**2)
    if ib==1:
        temp1=100.
        for j in range(ny):
            if tau_w.grid[ib][0,0,j,0]<temp1:
                i_min=j
                temp1=tau_w.grid[ib][0,0,j,0]
        temp1=dist[ib][i_min]
        dist[ib][:]-=temp1
        dist[ib-1][:]-=temp1
for ib in range(n_bl):
    plt.plot(tau_w.grid[ib][0,0,:,0],dist[ib])
print 'Leading edge index %i'%i_min    
```

```python
#mean tau_w
tau_mean=[]
for ib in range(n_bl):
    nx,ny,nz=tau_w.header[0][ib]    
    tau_mean.append(np.zeros([ny,4]))
    for k in range(nz):
        for j in range(ny):
            i=0
            for nit in range(len(steps)):
                tau_mean[-1][j,0]+=tau_w.data[ib][0,nit,0,j,k]
                tau_mean[-1][j,1]+=tau_w.data[ib][1,nit,0,j,k]
    for j in range(ny):
            tau_mean[-1][j,0:2]/=float(nz*len(steps))
    for k in range(nz):
        for j in range(ny):
            i=0
            for nit in range(len(steps)):
                tau_mean[-1][j,2]+=(tau_w.data[ib][0,nit,0,j,k]-tau_mean[-1][j,0])**2
                tau_mean[-1][j,3]+=(tau_w.data[ib][1,nit,0,j,k]-tau_mean[-1][j,1])**2

    tau_mean[-1][:,2:]/=float(nz*len(steps))   
    tau_mean[-1][:,2:]=np.sqrt(tau_mean[-1][:,2:])
```

```python
plt.figure()
for ib in range(n_bl):
    plt.plot(tau_w.grid[ib][0,0,:,0],-tau_mean[ib][:,0])
plt.figure()
for ib in range(n_bl):    
    plt.plot(tau_w.grid[ib][0,0,:,0],tau_mean[ib][:,2])        
plt.grid()
```

```python
corr_tau_qn=[]
for ib in range(n_bl):
    nx,ny,nz=tau_w.header[0][ib]
    corr_tau_qn.append(np.zeros([ny]))
    for k in range(nz):
        for j in range(ny):
            for nit in range(len(steps)):
                corr_tau_qn[ib][j]+=((tau_w.data[ib][0,nit,0,j,k]-tau_mean[ib][j,0])*
                                  (tau_w.data[ib][1,nit,0,j,k]-tau_mean[ib][j,1]))
    corr_tau_qn[ib][:]/=float(nz*len(steps))
    corr_tau_qn[ib][:]/=(tau_mean[ib][:,2]*tau_mean[ib][:,3])
```

```python
corr_tau_qn_2=[]
for ib in range(n_bl):
    nx,ny,nz=tau_w.header[0][ib]
    corr_tau_qn_2.append(np.zeros([ny]))
    for k in range(nz):
        for j in range(ny):
            for nit in range(len(steps)/2):
                corr_tau_qn_2[ib][j]+=((tau_w.data[ib][0,nit,0,j,k]-tau_mean[ib][j,0])*
                                  (tau_w.data[ib][1,nit,0,j,k]-tau_mean[ib][j,1]))
    corr_tau_qn_2[ib][:]/=float(nz*len(steps))
    corr_tau_qn_2[ib][:]/=(tau_mean[ib][:,2]*tau_mean[ib][:,3])
```

```python
plt.figure()
ids_=[1,0,2,3]
for ib in range(n_bl):
#    plt.plot(dist[ib],corr_tau_qn_2[ib]*2.)
    plt.plot(dist[ib],corr_tau_qn[ib],'b')
plt.grid()
plt.xlabel('$s$')
plt.ylabel(r'$r_{q_n,\tau_w}$')
plt.savefig('corr_tauw_qn.pdf')
tikz_save('corr_tauw_qn.tikz',figureheight='\\fh',figurewidth='\\fw')
```
