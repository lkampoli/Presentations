```python
def read_Latavg(filename,comp_TKE=False,ref_vel=None):
    store=False
    ifile=open(filename,'r')
    data_temp=[]
    varnames=[]
    for line in ifile:
        if line.strip().startswith('---') and store:
            store=False
            data_temp.append(entry)

        elif store:
            entry.append(float(line.split()[1]))

        elif line.strip().startswith('---') and not store:
            store=True
            entry=[]
            varnames.append(prev_line.split()[0])            
            
        prev_line=line
    if comp_TKE:
        if ref_vel==None:
            ref_vel=1.
        else:
            ref_vel=2./ref_vel**2
        data=np.empty([len(data_temp[0]),len(varnames)+1])
        for j in range(len(data_temp[0])):
            for i in range(len(varnames)):
                data[j,i]=data_temp[i][j]        
            data[j,-1]=0.5*(data[j,2]**2+data[j,3]**2+data[j,4]**2)*ref_vel
    else:
        data=np.empty([len(data_temp[0]),len(varnames)])
        for j in range(len(data_temp[0])):
            for i in range(len(varnames)):
                data[j,i]=data_temp[i][j]
    return data,varnames

def colNameToNum(name):
    import re
    pow = 1
    colNum = 0
    row=re.search('\d+', name).group()
    column=name[:-len(row)]
    row=int(row)
    for letter in column[::-1]:
            colNum += (int(letter, 36) -9) * pow
            pow *= 26
    return row,colNum

def excel_to_numpy(OPYXL,sheetname,range_excel,first_row_header=False,ax_chord_pos=None):
    """
    Methode returns a numpy array containing the real values of the cells
    
    Arguments
    OPYXL:              openpyxl library workbook (has to be opened before)
    sheetname:          sting containing the name of the worksheet
    range_excel:        sting with convetial excel range like "A1:C2" 
    first_row_header:   does the first row contain the title?
    
    returns:
    numpy array with data
    """
    ws=OPYXL[sheetname]
    #translate range information to length
    temp=range_excel.split(':')
    first_cell=colNameToNum(temp[0])
    last_cell=colNameToNum(temp[1])
    header=[]
    add_col=0
    if first_row_header:
        red_head=1
    else:        
        red_head=0        
        for i in range(last_cell[1]-first_cell[1]):
            header.append('col %i'%i)
    if not ax_chord_pos==None:
        add_col+=1
        ax_chord=1./float(ws[ax_chord_pos].value)
    ret_array=np.empty([last_cell[0]-first_cell[0]-red_head+1,last_cell[1]-first_cell[1]+1+add_col])
    if first_row_header:
        i=-1
    else:
        i=0
    for row in ws.iter_rows(range_excel):
        j=0
        if first_row_header and i==-1:
            for cell in row:
                header.append(cell.value)
            if not ax_chord_pos==None:
                header.append('ax_chord')
        else:
            for cell in row:    
                ret_array[i,j]=float(cell.value)
                j+=1         
            if not ax_chord_pos==None:
                ret_array[i,-1]=ret_array[i,0]*ax_chord
        i+=1
    return ret_array,header
```
