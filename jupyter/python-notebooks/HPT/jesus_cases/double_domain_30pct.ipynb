## Extract free stream line

```python
npoints=100
x_interp=np.linspace(-0.5,1.2,npoints)
scale_DNS=1.
y_hip=np.empty([npoints,2])
y_free=np.empty([npoints])
n_hip=len(S2L30_cascade.blade.data[:,0])    
diff=np.empty([npoints,2])
y_offset=0.006
for i in range(npoints):
    temp=[]
    if x_interp[i]<0.01:
        temp.append(0.)
        temp.append(0.)

    elif x_interp[i]> 0.95:
        temp.append(-.52-1.57/2)
        temp.append(-.52-1.57/2)    
    else:
        for j in range(n_hip-1):
            if ((S2L30_cascade.blade.data[j,0]*scale_DNS<=x_interp[i] and S2L30_cascade.blade.data[j+1,0]*scale_DNS>x_interp[i]) 
                or (S2L30_cascade.blade.data[j,0]*scale_DNS>=x_interp[i] and S2L30_cascade.blade.data[j+1,0]*scale_DNS<x_interp[i])):
                temp.append((S2L30_cascade.blade.data[j,1]*scale_DNS+(S2L30_cascade.blade.data[j+1,1]-S2L30_cascade.blade.data[j,1])/
                            (S2L30_cascade.blade.data[j+1,0]-S2L30_cascade.blade.data[j,0])*
                            (x_interp[i]-S2L30_cascade.blade.data[j,0]*scale_DNS))+y_offset)
    if len(temp)!=2:
        print 'found %i points in hip for index %i'%(len(temp),i)            
    if temp[0]<temp[1]:
        y_hip[i,0]=temp[0]
        y_hip[i,1]=temp[1]
    else:
        y_hip[i,0]=temp[1]
        y_hip[i,1]=temp[0]
    y_free[i]=0.5*(y_hip[i,1]+y_hip[i,0]+1.57)
```

```python
plt.figure()
plt.plot(x_interp[:],y_hip[:,0])
plt.plot(x_interp[:],y_hip[:,1])
plt.plot(x_interp[:],y_hip[:,0]+1.57)
plt.plot(x_interp[:],y_hip[:,1]+1.57)
plt.plot(x_interp,y_free)
```
